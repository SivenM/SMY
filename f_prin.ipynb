```
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def  trapP(X, x1, x2, x3, x4):
    """Функция строит трапециидальную функцию принадлежности П типа"""
    x = X
    y = []

    for i in range(0, len(x)):
        if x[i] <= x1 or x[i] >= x4:
            y.append(0)
        if x[i] <= x3 and x[i] >= x2:
            y.append(1)
        if x[i] < x2 and x[i] > x1:
            y.append((x[i] - x1) / (x2-x1))
        if x[i] < x4 and x[i] > x3:
            y.append((x4 - x[i]) / (x4 - x3))
   
    yy = np.array(y)
    #plt.plot(x,y,'-')
    #plt.grid()
    #plt.show()
    return yy

def treyg(X, x1,x2, x3):
    """функция строит треугольную функцию принадлежности   A  типа"""
    x =  X
    y = []
  
    for i in range(0, len(x)):
        if x[i] <= x1 or x[i] >= x3:
            y.append(0)
        if x[i] <= x2 and x[i] > x1:
            y.append((x[i]-x1)/(x2-x1))
        if x[i] < x3 and x[i] > x2:
            y.append((x3-x[i])/(x3-x2))
    
    yy = np.array(y)
    #plt.plot(x,y,'-')
    #plt.grid()
    #plt.show()
    return yy

def trapS(X, x1,x2):
    """Функция строит трапециидальную функцию s типа"""
    x = X
    y = []
  
    for i in range(0, len(x)):
        if x[i] <= x1:
            y.append(0)
        if x[i] >= x2:
            y.append(1)
        if x[i] > x1 and x[i] < x2:
            y.append((x[i]-x1)/(x2-x1))

    yy = np.array(y)
    #plt.plot(x,y,'-')
    #plt.grid()
    #plt.show()
    return yy

def trapZ(X, x1, x2):
    """Функиця строит трапециидальную функцию z типа """
    x = X
    y = []

    for i in range(0, len(x)):
        if x[i] <= x1:
            y.append(1)
        if x[i] >= x2:
            y.append(0)
        if x[i] > x1 and x[i] < x2:
            y.append((x2-x[i])/(x2-x1))

    yy = np.array(y)
    #plt.plot(x,y,'-')
    #plt.grid()
    #plt.show()
    return yy
```

```
# построим графики функций приндлежности П-типа, Лямбда-типа и Z-типа
X = np.linspace(0,10,1000)
f1 = trapP(X, 4, 6, 7, 9)
f2 = treyg(X, 2, 5, 8)
f3 = trapZ(X, 1, 3)
```

```
plt.plot(X, f1, '-')
plt.grid()
```

```
plt.plot(X, f2, '-')
plt.grid()
```

```
plt.plot(X, f3, '-')
plt.grid()
```

```
plt.plot(X, f1, '-')
plt.plot(X, f2, '-')
plt.plot(X, f3, '-')
plt.grid()
```

```
#композиция
f_comp = []
for i in range(0,len(X)):
    if f1[i]>=f2[i] and f1[i]>=f3[i]:
        f_comp.append(f1[i])
    elif f2[i]>=f1[i] and f2[i]>=f3[i]:
        f_comp.append(f2[i])
    else:
        f_comp.append(f3[i])

plt.plot(X, f_comp, '-')
plt.grid()
```

```
# центр тяжести
c1 = np.trapz(X*f_comp, x=X)
c2 = np.trapz(f_comp, x=X)
x_center = c1/c2
print("Центр масс по х равен ", np.around(x_center, decimals=2))
```
